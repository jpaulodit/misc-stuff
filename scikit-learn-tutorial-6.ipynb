```python
sns.pairplot(data, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', size=7, kind='reg')
```

```python
# Use linear regression to estimate

# create a python list of feature names
feature_cols = ['TV', 'Radio', 'Newspaper']

# use the lsit to select a subset of the original dataframe
X = data[feature_cols]

# equivlent command to above
X = data[['TV', 'Radio', 'Newspaper']]

X.head()
```

```python
y = data['Sales']

y.head()

print type(y)
print y.shape
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
```

```python
linear.fit(X_train, y_train)
```

```python
print linear.coef_
print linear.intercept_
```

```python
zip(feature_cols, linear.coef_)
```

```python
# make the prediction
y_pred = linear.predict(X_test)

print y_pred
```

```python
# We see from above that the Newspaper feature has little correlation with sales. Let try to remove it and see how
# our model fares. Note if we need to compare how well each model does against the other, we need to set a random_state
# when splitting the test data to keep it consistent.

X = data[['TV', 'Radio']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

linear = LinearRegression()

linear.fit(X_train, y_train)

y_pred = linear.predict(X_test)

np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```
